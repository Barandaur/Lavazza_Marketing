# A closer look at the product

While the general idea of our product is pretty clear by now, there are a lot of details to be discussed when it comes down to prototyping. How many tastes and combination to create? How many instruments to include in each kit? Should all kits include such instruments? Should we create kits for each of our personas?

These questions highlight some general potential pain-points in our kits:  the same persona, may need different kits, whether for taste preferences, number of servings, or because nobody wants to buy accessories already owned, having even to get rid of them, on top of environmental issues. 
Thus, not all kits should include such accessories in them, differentiating among “First time kit” and “Re-fill kit”, where the latter contains only the ingredients.

At the same time, kits may need to be different also across personas; for example Marta may want an analcoholic kit if she wants to use our product in the mid-afternoon, as a break from studying, while Arianna may specifically want a tsa-approved kit.

By doing some quick maths, we see that *we would need a number of kits which is really hard to sell in retail*; this means that either one needs to investigate a lot of personas, find some common traits and sell just few kits,compatible with a decently broad audience, or other distribution channels have to be preferred, such as the online one. 

We will further discuss this issue in the next sections, when talking about our proposed strategy. Assuming, for now, that the **online distribution** solves this problem, another aspect must be taken into consideration: **packaging**. When we anylized competitors for Arianna's travelling customer journey, we have seen that spilling or damaged products can be a huge pain-point for our clients; thus, packaging should be optimized with regard to this issue. At the same time, packaging is part of the customer experience for these kind of products, so it should convey ideas of elegance and luxury, factors we considered important enough for our product to build positioning maps on them. These two aspects can be hard to combine, excpecially since we have large combinations of different types of kits.

One way to account for this is to standardize a lot of things, like the dimensions of the bottles, so that packages for the same number of servings can be the same, and can be designed well with regard to the two aspects previously highlighted. In this way, customer may also have a decent degree of customization, as, based on the number of servings they want, they can choose whether to buy a pre-designed kit or to choose by themselves what to include for each component, e.g. what alchoolic, what sweetener etc. He can also choose whether to include or not accesories, such as a shaker; if one chooses too many accessories, that couldn't be elegantly fit into a single kit, 2 different packages are delivered.

As we can see, **customization** is becoming a possibility, and Lavazza may change the package strategy according to the degree of customization; i.e. for the highest degree of customization one would need to care less about perfectly fitting products into pre-dseigned compartments, opting instead for putting them into bags inside the kit, or just opting for a bigger package in which, on top of a confetioned kit, there are also other products the customer asked for. This last possibility would allow customers to choose not only the particular instance of each kit component (e.g. which alchoolic, if any, among those that would fit for that number of servings), but also to buy products that wouldn't otherwise fit, e.g. several sweeteners instead of just one, or an alchoolic for a different number of servings. This would allow for a high degree of customization, while keeping the kit package design (actually a customer may even decide he doesn't want the kit package, if he prioritizes having less waste over luxury), although putting it into a bigger package together with other products, thus making the initial unboxing experience less 'luxorious'. Since customization is becoming a parameter for the general strategy, we will run an analysis on it in the next section


We reasoned about the possibility of consuming the product at two different times of the day which imply different **recipes**: one that adapts to the aperitif while the other to a post dinner situation.
For the sake of prototyping, we asked about possible recipes to friend of ours, who is a reasonably famous bartender/cocktail maker. 

For the aperitif kit, the cocktail to be prepared could be a shaken Campari (or a similar product) with a twist on the coffee: the preparation requires a short infusion with Lavazza coffee beans in the Campari, and then a shake with the red vermouth and ice.

# Customization analysis

We ran an analysis on customization by brand. Notably, we will consider the percentage of followers of several brands that customize their own profiles on Twitter (changing both profile image and background settings). Of course, results may be biased by the fact that some users are less able or willing to go through the settings, and it is reasonable to think that this happens more the older the user is. Thus, if a brand only attracts older people, it may seem that its customers care less about customization; it isn't necessarily false, but it would be hard to infer, due to the spurios correlation. More in general, this metric is valid in a digital framwork, this may carry a bit of bias if we use results to generalize to an analogic scenario. Hopefully, since, older people are also less likely to use Twitter (Zagheni et al. (2017) research backs this up, altough it refers to another country and Facebook usage), and since customising one's background on Twitter is extremely easy and intuitive, this confounding doesn't impact too much results of the analysis. 


One way to check for the relevance of this bias (using Twitter data) would be to build a big enough dataset, keep the users who include age in their description, get other information on them, like the page liked, and build a ML algorithm that allows to classify into classes of age based on such information. If this algortihm does its job well, one can check for the correlation between age and customization in this setting.


Another possible objection to the analysis is that some brands may have more followers that use Twitter less, and, if a follower doesn't use Twitter, he has less chances of customising its account (it may even be a bot). We will do a robustness check for this behavour, which will consist in keeping only users who have published a minimum number of tweets *or* follow a minimum number of pages, considering these 2 parameters as a proxy for a minimum degree of Twittare usage.


Keeping the first mentioned limit of the analysis in mind, results seem to have face validity, as brands like MonsterEnergy and Starbucks score higher on the percentage of customizing users, and so do brands about recipes sharing.

```python
def customization_by_page(df, twitter_page, both_img_theme = False, *threshold):
    '''
    Purpose: given a df, calculate how many followers of a page changed 
             profile image and/or background and return the ratio.
             
             If a keyword argument is given, it will only consider users who
             have published enough tweets or have enough followed pages, to 
             avoid the risk of underestimating the ratio due to people who
             don't customise because they don't use twitter much
    '''
    
    df = df[df['twitter_page'] == "{}".format(twitter_page)] 
    num_of_follw = len(df)
   # print("{} followers in the df".format(num_of_follw))
    
    if threshold:
        print("considering only users who have used twitter enough")
        df = df[(df['status_count'] >= threshold[0][0]) | (df['friend_count'] >= threshold[0][1])]
    
    if both_img_theme == True:
        df = df[(df["changed_theme"] == 1.0) & (df["changed_img"] == 1.0) & (df["twitter_page"] == "{}".format(twitter_page))]
        print("considering followers who changed both image and theme")
    else:
        df = df[(df["changed_theme"] == 1.0) | (df["changed_img"] == 1.0) & (df["twitter_page"] == "{}".format(twitter_page))]
        print("considering followers who changed at least one between image and theme")
        
    num_of_customising_follw = len(df)
    #print("{} followers have customised their twitter".format(num_of_customising_follw))
    ratio = num_of_customising_follw / num_of_follw
    print("{}% of {} users have customised their twitter \n".format(round(ratio,3), twitter_page))
    
    return df, ratio    


# for plotting results
def plot(list_w_dict, col = 'green'):
    sb.set_context('poster') # use large font
    plt.figure(figsize=(10, 10))
    
    dff = pd.DataFrame.from_dict(list_w_dict[0], orient = 'index').sort_values(by =0)
    descriptors = list(dff.columns.values)

    plt.subplot(131)
    plt.barh(dff.index.values, dff[descriptors][0].to_list(), color = col)
    if len(list_w_dict) > 1:
        
        dff2 = pd.DataFrame.from_dict(list_w_dict[1], orient = 'index').sort_values(by =0)
        plt.subplot(131)
        plt.barh(dff2.index.values, dff2[descriptors][0].to_list(), color = 'green')
    plt.show()
    
    
    
def twitter_usage_by_page(df, twitter_page,  threshold):
    '''
    Purpose: given a df, calculate how many followers of a page follow at least
             a number of pages OR have published a number of tweets

    ''' 
    df = df[df['twitter_page'] == "{}".format(twitter_page)]     
    df2 = df[(df['status_count'] >= threshold[0]) | (df['friend_count'] >= threshold[1])]

    ratio = len(df2) / len(df)
    print("{}% of {} users are above the thresholds \n".format(round(ratio,3), twitter_page))
    
    return df, ratio    
```

#### Clout-Vulnerability matrix

```python
# Loading functions used

# Clout: how much power brand A has over brand B
def clout(A,B):
        inters = len(A.intersection(B))
        clout = inters/len(B)
        return clout

# Vulnerability: how vulnerable A is to B's strategic moves
def vulner(A,B):
        inters = len(A.intersection(B))
        vulner = inters/len(A)
        return vulner
    

def clout_df(d = {} , prints=False, check_vulner=False):
    
    """Function that computes C-V matrix and returns it. 
    prints is for debugging and have asense of prograssion.
    check_vulner checks for correctness of computations/formulas."""
    
    CLOUT= {}  #instatiate an empty dictionary
    for p in d:  
        CLOUT["%s" %p]= {} #instatiate an empty sub-dictionary CLOUT["%s" %p] inside CLOUT

        for g in d:  #we now compute the clout of p over each g brand

            #inside the sub-dictionary CLOUT["%s" %p], we add a key with the name of the g brand
            CLOUT["%s" %p]["%s" %g]=clout(d["%s" %p],d["%s" %g])
            
        if prints: print (p, "clout computed") # debug printing single brand clout completed

    if prints: print (CLOUT)  #CLOUT nested dictionary!
    
    # This part checks if computations are done correctly by computing the vulnerablility matrix 
    # and confront it with the transpose of the clout
    if check_vulner:
        VULNER= {}  
        for p in d:  
            VULNER["%s" %p]= {} 
            for g in d:
                VULNER["%s" %p]["%s" %g]=vulner(d["%s" %p],d["%s" %g])


    return pd.DataFrame(CLOUT).add_prefix('Clout ')
```

```python
brands=["lavazzagroup","illyIT","Kimbo_It","CaffeVergnano","aperolspritzita","CampariItalia","BACARDI_IT", 
       "MonsterEnergyIT", "redbullITA"]

brands_followers={}
for m in brands:
    brands_followers['%s' %m]=set(cur.execute("SELECT followerid FROM twitter_data WHERE brand='%s'" %m))
    
clout_df(brands_followers)
# Vertical is clout and horizontal is vulnerability
```

[Back on top](#top) 


Once we will have had success with Andrea and we will have obtained a spot in the cocktail market, we could start thinking to satisfy other market segments such as Marta's and Arianna’s.

**Marta** is a student, and in the coffee consumption she is primary driven by a need for energy and a need for aggregation. She drinks coffee before (or during breakes from) studying, or when she is with friends. She is more functionally driven, compared to someone like Andrea. Our product would allow her to organize some pre/post dinner moments with her friends and to satisfy the status seeking she is looking for, by preparing and having something good looking in her glass. At the same time, she may just want to chill, unwilling to pursue a preparation which is complicated and time consuming.
For this reason, somewhat different kits has to be developed in order to make our product suitable also for her.


A possible solution consists in launching two different **‘student edition’ kits** (with more to add at later stges) to sell also at the **supermarket**, one for an aperitif cocktail and the other for a post-dinner drink. Each will require very few preparation steps, including pre-batch cocktails to which to only add some coffee, without the need of any accessory, aside, possibly, from a shaker. In this way, we will provide students with an affordable, easy to realize product that still promises good looking results to be posted on Instagram, while still allowing them to explore other possibilities online. 



This will also help in avoiding to propose too many kits in retail. At the same time, expanding to the supermarket can be a good idea to reach her. Indeed, students tends to arrange their evening plan at the last minute.
If one's friends are going to arrive in an hour, going quickly to the supermarket is still possible, making an order on the Lavazza website and receive the kit immediately is not. 
However, when student go to the supermarket, they know exactly what they need and the location of the products they are looking for. They don’t usually ‘explore’ the shelves in search of new inspiring products. Thus, in order for our kit to be successful, it is best if students know about it before going to the supermarket. 



To expose them to the kit, it is fine to use social media, but also more traditional means can be effective. For example, if, on top of Marta, Lavazza wants to target students who enjoy to relax watching reality shows (they can be a good match for an easy but tasty cocktail to consume alone or with house-mates), then TV commercials can be ideal as well.
Aside from these possibilities and focusing on Marta, social media can work well, possibly with collaborations with other brands, such as Aperol if Lavazza will take this path. Another possibility would be to create an event or a situation that Martas will tend to share on their own, allowing Lavazza to indirectly reach their peers, rather than with an explicit commercial. Roughly, this is the idea behind some forms of ambient marketing, which can be a winning strategy to reach Marta. 



Regarding **Arianna**, she is a traveller and, as we highlighted before, she cannot bring alcohol in the airport. Similarly to Marta, Arianna needs a simple recipe because, during a flight, on the small airplane table, complicated preparation procedures are unfeasible.
Hence the traveller kits will be similar to those developed for students with the main difference being that the alcoholic beverage will be actually alcohol-free and the kit will also provide a sachet of instant coffee. A bar spoon will be included as well, since a traveller cannot have his/her own kitchen utensils with himself/herself. 


The traveller kit will also differ from the student one in term of packaging. A traveller mostly needs one-person doses and something that is comfortable to carry, meaning that in this case the kit will be smaller.
For the initial marketing campaign, it is possible to make use of the already established collaboration with Luis. While he is travelling around the world for the ‘Breakfast Hub’ project, he might make additional videos where he prepares some cocktails with our kits, capturing on camera the curiosity of the other travellers. 
At the end of the video he refers to the kits if one wants to try himself, mentioning also a giveaway, to help spread the news and to boost the sentiment of those who will receive it.


From the analysis carryed out in 1.0 (section pages_liked), we found out that more than 1 out of 4 Lavazza's followers travellers care a lot about luxorious journeys. This is good news for us, but we have to handle this segment well; indeed, he may not be interested or even reached by an advertising with Luis. Hopefully, though, he will be reached by the advertisement made for Andrea, as he seems to be a blend of these two personas. One could collaborate with Alitalia, that appears to be followed by a significant share of these personas, to specifically reach these customers in an elegant way, by potentially offering the kit not before the travel but during it, solving all the problems of making it tsa-approved. 



Finally, another idea, tailored to Arianna and Marta, is to organize an Instagram **challenge** where participants are asked to prepare their own take of a coffee cocktail and post a picture of it on their social pages. The recipe of the cocktail receiving more likes will be added to the online catalogue and will be given the name of the inventor. It is needed to consider this idea with care though, as an active social media marketing can [backfire](https://www.metro.us/mcdstories-mcdonalds-gets-fried-in-marketing-backfire/).
If Lavazza deems worth to make this challenge, it can be used at the beginning, to help spread the knowledge of the kit's existance. Alternatively, one can opt for this challenge at a later stage, possibly to build a virtual community around a dedicated Lavazza **app**. Indeed, when there will be enough kits and people using them, Lavazza may want to launch an app specifically dedicated to cocktail kits, which can make it easier to users to navigate, customize, rate, and, possibly, discuss or share their works in a socialmedia-like feed

# 4.1 Privacy considerations

Privacy is a big topic nowadays, to tackle its relation with out product, we must first make a small digression on the data our kits will make possible to retrieve, first considering what data Lavazza may want to gather and why.


Since our main distribution channel will be online, there are a number of information that Lavazza can gain from the purchases of our product, more so since it is more convenient to finalise the purchase upon logging in.
First of all, since there will be many types of available cocktail kits, differing among key ingredients, it could be possible to build a recommendation system, based on what info the user provides on his profile and on his customer behaviour compared to other customers ones. Furthermore, this could allow also to understand better which kind of tastes are more appealing for a customer, potentially helping in the creation of new beverages or new cocktails, beneficial to the Flagship store as well. 


Building up on this possibility, Lavazza could also launch new cocktails as kits to test how they are received by the market, since it can more easily obtain feedback online, thanks to tools like the review system. All of these possibilities would be more powerful if Lavazza manages to create an online community around this product, especially after the app release. Indeed, some customers may want to gather online for several reasons, for example to share their personal takes on the kits - more so if the customization aspect is emphasized - or to feel like belonging to a more ‘sophisticated’ niche of coffee users,  or just to express themselves, if Lavazza manages to create a space for them to do so. Not only this would boost product usage, but it would also provide access to more feedback and information, for example on when and how the product is consumed. 


Similarly, connecting to the page with the instructions on how to make a cocktail would provide access to the aforementioned piece of information on when the product is used, allowing Lavazza to see if the kits are used in the moments of the day they were designed for. Indeed, while one may check the online recipe also much before making the cocktail, actually scanning the code inside the package is more likely to happen right before consumption.
Data from the website/app could also allow to understand how frequently different customers purchase the product, which, combined with their personal information and the moment of the day in which they use it, could allow to better understand their drivers for the purchase; e.g., do they buy the product regularly and use it always around the same time, or do they buy it more frequently around some periods? Do they buy it for themselves or do they also request for a gift enveloping? 
In the latter case, it could be interesting to have information on the conversion rate of users who received the product as a gift. An idea to track this behaviour is to ask, upon registration, if the first time one has consumed the product was thanks to a gift. Furthermore, to incentivise the app usage and to potentially track who gave a gift to whom, people could have the possibility of inviting friends over the app, with the reward being a small discount.


Thus, if one had access to such data, on top of what mentioned, he could know addresses, frequency of usage of kits (thus inferring on income level or potential addiction), one may have an hint about whether a person is about to travel (if he buys a travel kit without selecting the gift option). Furthermore, it could use data about when the user checks a recipe or logs the website to try to reconstruct his day and his moment of usage. Of course, Lavazza can use these data to ideate new products, hopefully keeping ethical issues in mind, but there is some sensitive information that needs to be protected

# 4.2 Ethical considerations

Up to some years ago, ethical aspects didn't own their place in a business report. Though, with the rise of the importance of CSR, as well as with the change of some managerial paradigms, notably transitioning from the idea of just making profit to that of gaining prestige, things are slowly changing. In accordance with such changes, we will briefly talk about potential ethical issues of our product.


Alchool and coffee aren't two of the healthiest products, combining them can help alleviate the shortcomings of each, but it can boost them as well. An example of the former is that some people may satisfy their coffee needs with one our kits, which, potentially, can contain a smaller dose of caffeine. This would be particularly relevant, as the product is thought for an use in the afternoon/evening, when some people may already have drunk several coffees; while general research on benefits and harms of coffee is inconclusive, it is certain that a too large daily coffee intake can lead to major health issues.
At the same time, though, the opposite of the scenario we just depicted can be true: some people that before would have consumed a simple aperitif beverage, or a sweet or even nothin, may now be attracted by this new product, thus increasing their coffee and alchool intake for that day. 
Despite this possibility, there are some good news: our product isn't exactly cheap, and it is thought to be used on occasions, so, if we focus on the average usage of our product, we can say that it is reasonably harmless.


Things aren't the same, though, if we focus on other aspects as well. Indeed, our product may bring some people closer to the world of alchoolics and coffee, leading, possibly, to addiction. Since a relevant part of our campaign is carryed out through online means, though, we can take advantage of technological advancments. One idea could be to collaborate with social media owners to identify and exlude from our advertisings people who are proner to incur into depression or addiction, using an NLP-based classifier. While it may seem an unfeasible idea, such technologies are already existant and in use, expecially from big brands like Facebook, so it's just a matter of putting them to good use, by specifically addressing this issue with social media companies.


Finally, our product may bring more people to consume alchool and/or coffee during the evening, deterioring the quality of their sleep. While this doesn't generally seem like a big problem, it can become a matter of relevance if we consider those subjects who already come from a deteriorated mental state, for whom quality of sleep is particularly important to give them the energy and strenght to fight their battle. Thus, similar considerations to the above apply.


One realistic opportunity Lavazza has for doing CSR on this potential problems, would be to make donations to mental health organizations, advertising it in a campaign in which it disencourages eccessive use of its own products, promoting a sustainable coffee culture.
